```python
import numpy as np
```

```python
#Create matrix
A = np.array([[1,2,3],
              [4,5,6]])
B = np.array([[7,8],
             [9,10],
             [11,12]])
print("Matrix A:\n", A)
print("Matrix B:\n", B)
```

```python
#Matrix Transpose
print("Tranpose of A:\n", A.T)
```

```python
#Dot product (Matrix multiplication)
result = np.dot(A,B)
print("Dot product (A x B):\n", result)
```

```python
#Element-wise Multiplication
C = np.array([[1,2], [3,4]])
D = np.array([[5,6], [7,8]])

print("Element-wise multiplication:\n", C*D)
```
