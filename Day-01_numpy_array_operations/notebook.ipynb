```python
#Importing the numpy library
import numpy as np
```

```python
#Create arrays
arr1 = np.array([1,2,3,4,5])
arr2 = np.arange(10)
arr3 = np.zeros((3,3))
arr4 = np.ones((2,4))

print(arr1)
print(arr2)
print(arr3)
print(arr4)
```

```python
#Basic operations
print("Addition:", arr1 +5)
print("Multiplication:", arr1 * 2)
print("Square:", arr1**2)
```

```python
#Statistical Operations
print("Mean:", np.mean(arr1))
print("Sum:", np.sum(arr1))
print("Max:", np.max(arr1))
print("Min:", np.min(arr1))
```

```python
#Reshape
reshaped = np.arange(12).reshape(3,4)
print(reshaped)
```

```python
#Indexing & Slicing
print(arr1[2])
print(arr1[1:4])
```
